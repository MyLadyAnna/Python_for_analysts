### 1.1. Скачать файл kc_house_data.csv - приложен на платформе

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('kc_house_data.csv', sep = ',', encoding='latin-1')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.hist(df['price'])   
plt.xlabel('Стоимость')              
plt.ylabel('Количество')           
plt.title('Стоимость недвижимости'); 
```

```python
sns.histplot(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во');
```

Большую часть данных составляет недвижимость стоимостью от 0.1 до 0.8 млн.  
На графике из библиотеки **seaborn** также видно, что есть очень малое количество домов стоимость около 7.8 млн, 7.1 млн, 6.9 млн и от 4 до 5.8 млн. Самое большое количество домов примерно 1170 стоит около 0.25 млн. 

### 1.3 Изучите распределение квадратуры жилой

```python
sns.histplot(df['sqft_living'])   
plt.xlabel('Квадратура')              
plt.ylabel('Количество')           
plt.title('Квадратура недвижимости'); 
```

Большая часть недвижимости в данных - это дома с жилой квадратурой от 1 до 3 тыс. Самое большое количество домов имеют квадратуру - 2тыс. Есть очень малое количество домов с квадратурой выше 8 тыс

### 1.4 Изучите распределение года постройки

```python
sns.histplot(df['yr_built'])   
plt.xlabel('Год постройки')              
plt.ylabel('Количество')           
plt.title('Года построек недвижимости');
```

Больше всего домов были построены примерно в 2005 (примерно 1550). Среди данных нет домов, которые были бы построены до 1900 года или после 2018. Наименьшее количество домов было ппостроенов 1935 (около 100)

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix,1)
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
```

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

Чем выше grade оценка, тем выше цена дома. Самые дорогие дома имеют оценку 13.

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg');
```

На цену влияет и квадратура чердака. Наибольшую стоимость имеют дома с наибольшей квадратурой чердака.

```python
sns.boxplot(x=df['view'], y=df['price'], whis=1.5)

plt.xlabel('Оценка вида')
plt.ylabel('Цена')
plt.title('Зависимость цены от вида')
plt.xticks(rotation=60);
```

Оценка вида из дома также влияет на его цену. Цена выше, если оценка вида выше. У домов с более высокой оценкой минимальная цена также выше.

```python
sns.boxplot(x=df['bathrooms'], y=df['price'], whis=1.5)

plt.xlabel('Кол-во ванных комнат')
plt.ylabel('Цена')
plt.title('Зависимость цены от кол-ва ванных комнат')
plt.xticks(rotation=60);
```

Кол-во ванных комнат также влияет на цену. В основном чем больше ванных комнат тем выше стоимость дома.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

Жилая площадь влияет на цену дома. Чем больше жилая площадь, тем выше стомость жилья.
